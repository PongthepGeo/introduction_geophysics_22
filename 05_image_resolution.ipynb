#### Image resolution

Geophysical images and general images (photo) resolutions have several common properties. This exercise compares high and low resolutions of the giraffe picture, which shows a big picture of limitations in geophysics imaging; <font color="#EAC7B9"> we did not feel fine subsurface texture, but we felt just rough skin </font>.   

```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from scipy.ndimage import gaussian_filter
```

```python
image = Image.open('datasets/giraffe.jpg')
image = np.asarray(image)
image = image[:, :, 0] # show only the first channel
print(image.shape)
blur = gaussian_filter(image, sigma=10)
```

```python
ROW, COL = image.shape

fig = plt.figure(figsize=(20, 10))

ax1 = fig.add_subplot(1, 4, 1)
# to draw a line from (205, 0) to (205, 720 )
x = [205, 205]
y = [0, ROW]
ax1.plot(x, y, color='red', linewidth=3, linestyle='--')
ax1.imshow(image, aspect='auto')
ax1.title.set_text('Original image')
ax1.set_xlabel('pixel-x')
ax1.set_ylabel('pixel-y')

ax2 = fig.add_subplot(1, 4, 2)
axis_y = np.linspace(0, ROW, ROW)
ax2.plot(image[:, 205], axis_y, color='red', linewidth=1, linestyle='-')
ax2.title.set_text('Signal at column 205')
ax2.invert_yaxis()
ax2.set_xlabel('amplitude')
ax2.set_ylabel('pixel-y')

ax3 = fig.add_subplot(1, 4, 3)
ax3.plot(x, y, color='pink', linewidth=3, linestyle='--')
ax3.imshow(blur, aspect='auto')
ax3.title.set_text('Blur image')
ax3.set_xlabel('pixel-x')
ax3.set_ylabel('pixel-y')

ax4 = fig.add_subplot(1, 4, 4)
axis_y = np.linspace(0, ROW, ROW)
ax4.plot(blur[:, 205], axis_y, color='pink', linewidth=1, linestyle='-')
ax4.title.set_text('Signal at column 205')
ax4.invert_yaxis()
ax4.set_xlabel('amplitude')
ax4.set_ylabel('pixel-y')
```

From a quick look, we recognize that the cross-section of column 205 in the blurred image shows a lower frequency than the original image, and this kind of signal is typical to be observed in most geophysical subsurface imaging. In a nutshell, the earth's materials absorb (attenuation) most energy sources causing the return signal (reflection) to have weaker energy (low frequency).
