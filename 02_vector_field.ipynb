#### Vector field

In geophysics, scalar, vector, matrix, and gradient are related. This exercise demonstrates a variation between weights and a number of drinking bubble tea in which calculation a vector field.

First, observing weight and number of bubble tea as a function of $f(x, y)$, e.g., $x = 2.2$ and $y = 50.8$. This data point reads as drinking 2.2 cups, weighing 50.8 kg.

Often, observation in geophysics contains multiple observed stations that record recurring earth phenomena, e.g., temperatures. Considering one station, we will have a scalar datum, and several stations will have data in a matrix form.   

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = 2.2, 50.8
fig, ax = plt.subplots(figsize=(10, 10))
plt.scatter(x, y, s=50, color='orange', marker='o', edgecolors='black')
```

The observation will continue and we record the second data point as $f(x, y) \Rightarrow f(1.8, 50.2)$. From these two observed data points can use to compute vector field. 

```python
import math 

def mag(x): 
    return math.sqrt(sum(i**2 for i in x))

day_1 = np.array([2.2, 50.8])
day_2 = np.array([1.8, 50.2])
length_x = day_2[0] - day_1[0]
length_y = day_2[1] - day_1[1]
vector_PQ = np.hstack((length_x, length_y))
print('magnitude of vector PQ: ', mag(vector_PQ))

fig, ax = plt.subplots(figsize=(10, 10))
plt.quiver(day_1[0], day_1[1], length_x, length_y, color='red', angles='xy', scale_units='xy', scale=1)
plt.scatter(day_1[0], day_1[1], s=500, color='orange', edgecolor='black', linewidths=5, alpha=0.2)
plt.scatter(day_2[0], day_2[1], s=500, color='orange', edgecolor='black', linewidths=5, alpha=0.2)
plt.xlim(0, 4)
plt.ylim(46, 54)
plt.title('vector of two observed points')
plt.grid(color='green', linestyle='--', linewidth=0.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x,y = np.meshgrid(np.linspace(-5, 5, 10),np.linspace(-5, 5, 10))

u = -y/np.sqrt(x**2 + y**2)
v = x/np.sqrt(x**2 + y**2)

fig, ax = plt.subplots(figsize=(10, 10))
plt.scatter(x, y, s=50, color='red', marker='*')
plt.quiver(x,y,u,v, color='blue')
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')
```
