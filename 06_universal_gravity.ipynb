#### Universal gravity


\begin{align*}
F & =\frac{G\times m_{gold} \times m_{earth}}{r^{2}}\\
F & =m_{gold} \times a\\
m_{gold} \times a & =\frac{G\times m_{gold} \times m_{earth}}{r^{2}}\\
g & =\frac{G\times m_{earth}}{r^{2}}
\end{align*}


\begin{align*}
F: & attractive\ force\ ( N)\\
G: & universal\ gravitational\ constant\ \left( 6.674\times 10^{-11} ,\ m^{3} \cdot kg^{-1} \cdot s^{-2}\right)\\
m_{gold} : & mass\ of\ smaller\ object\ ( kg)\\
m_{earth} : & mass\ of\ larger\ object\ ( kg)\\
r : & distance\ between\ the\ two\ masses\ ( m)\\
a: & gravitational\ acceleration\ \left( m\cdotp s^{-2}\right)\\
g: & geophysical\ exploration\ expresses\ a\ as\ g\ \left( m\cdotp s^{-2}\right)
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = 6.674*pow(10, -11)
m = 5.972*pow(10, 24)
meter = pow(10, 6)
r = np.array([6.3781*meter, 6.3881*meter, 6.4781*meter, 100*meter])
f_1 = (G*m)/pow(r, 2)
dis = np.linspace(6.3*meter, 100*meter, 100)
f_2 = (G*m)/pow(dis, 2)
```

```python
_, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5), tight_layout=True)
ax.ticklabel_format(useOffset=False, style='plain')
plt.scatter(r, f_1, s=500, c='red', marker='o', edgecolor='black')
plt.plot(dis, f_2, color='orange')
plt.title('Universal Gravity', size=18, fontweight='bold')
plt.xlabel(r'$distance\ (m)$')
plt.ylabel(r'$gravity\ acceleration\ (m\cdotp s^{-2})$')
```
