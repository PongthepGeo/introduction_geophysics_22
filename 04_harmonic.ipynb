#### Simple harmonic

Suppose we would like to measure a student's height, and this case is simple measuring by using a ruler because the tool and the object are physically linear. If we want to measure wave phenomena caused by droplets, the ruler might not be suitable.

Harmonic function is an alternative solution that can transfer wave geometries into understandable measurements.

$$\displaystyle y( t) =a\cdotp cos( \omega \cdotp t+\theta )$$


\begin{align*}
a: & amplitude & \theta: & phase\ angle\\
\omega: & angular\ frequency & T: & period\\
\omega: & \frac{2\pi }{T} =2\pi f & f: & frequency\\
t: & time & \lambda: & wavelength
\end{align*}

Keep in mind that np.sin($\theta$) uses radian as an input. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sin_cos(amplitude, frequency, time, phase, choice):
	'''
	np.cos and np.sin require input as radians (not degrees)
	'''
	y = 0 # preallocate memory
	if choice == 'cos':
		y = amplitude*np.cos(2*np.pi*frequency*time + phase)
	elif choice == 'sin':
		y = amplitude*np.sin(2*np.pi*frequency*time + phase)
	return y
```

```python
amplitude = 2  # no unit
frequency = 0.5 # cycle per second (second-1)
time = np.linspace(0, 2*np.pi, 180) # second
phase = np.pi/2 # where to start (radians)

y = sin_cos(amplitude, frequency, time, phase, 'sin')
x = np.linspace(0, 360, len(y)) # degree
plt.plot(x, y, color='blue')
plt.xlabel(r'degree ($\degree$)')
plt.ylabel('amplitude')
plt.title('Sine')
```
