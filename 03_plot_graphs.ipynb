#### Plot graphs

This exercise demonstrates how to plot graphs using **pandas** library.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##### 1. Plot function using matplotlib 

```python
data = pd.read_csv('datasets/magnetics.csv')
print(data.describe()) # check data statistics
```

```python
plt.plot(data['Bx'], color='red', linestyle='--', label='Bx')
plt.plot(data['By'], color='blue', linestyle='--', label='By')
plt.plot(data['Bz'], color='green', linestyle='--', label='Bz')
plt.plot(data['BT'], color='black', linestyle='-', label='BT')
plt.legend(loc='upper right')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Magnetics')
```

##### 2. Median

Middle value separating the greater and lesser halves of a data set.

```python
print(data['BT'].median())
```
